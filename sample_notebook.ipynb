## Data Analysis Example

Let's create some sample data and analyze it.

```python
# Create sample data
data = {
    'Product': ['Bond A', 'Bond B', 'Bond C', 'Bond D'],
    'Yield': [2.5, 3.1, 2.8, 3.4],
    'Price': [98.5, 102.3, 99.8, 101.5]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Calculate statistics
print(f"Average Yield: {df['Yield'].mean():.2f}%")
print(f"Average Price: {df['Price'].mean():.2f}")
```
